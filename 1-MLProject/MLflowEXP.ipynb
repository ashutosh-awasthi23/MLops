#### Important Libraries Import

```python
import pandas as pd
import kagglehub
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

#### EDA

```python

path = kagglehub.dataset_download("markmedhat/titanic")
print("Path to dataset files:", path)
```

```python
data_frame = pd.read_csv(r"C:\Users\ashut\.cache\kagglehub\datasets\markmedhat\titanic\versions\1\titanic.csv")
data_frame
```

```python
data_frame.info()
```

```python
data_frame.drop(columns=["PassengerId"],axis=1,inplace = True)
#It returns None
```

```python
data_frame
```

```python
data_frame.drop(columns= ["Name","Ticket"],axis=1,inplace =True)
```

```python
data_frame
```

```python
data_frame["Embarked"].unique()
```

```python
df = data_frame.copy()
```

```python
df.dropna(subset = ["Embarked"],inplace  = True)
```

```python
df.drop(columns=["Cabin"],axis=1,inplace = True)
```

```python
le = LabelEncoder()
ohe  = OneHotEncoder(sparse_output = False)
```

```python
data_frame["Embarked"].nunique()
```

```python
encoded = ohe.fit_transform(data_frame[["Embarked"]])
encoded_df = pd.DataFrame(encoded,columns= ohe.get_feature_names_out(["Embarked"]))
```

```python
encoded_df
```

```python
real_df = pd.concat([data_frame,encoded_df],axis=1)
```

```python
real_df.dropna(subset = ["Embarked"],inplace=True)
```

```python
real_df.drop(["Embarked"],axis=1,inplace=True)
```

```python
real_df.drop(["Embarked_nan"],axis=1,inplace =True)
```

```python
real_df.drop(["Cabin"],axis=1,inplace = True)
```

```python
real_df
```

```python
real_df[real_df["Age"].isna()]
```

```python
real_df["Age"] = real_df.groupby("Sex")["Age"].transform(lambda x : x.fillna(x.mean()))
```

```python
real_df
```

```python
real_df["Sex"] = le.fit_transform(real_df["Sex"])
real_df
```

```python
y = real_df["Survived"]
X = real_df.drop(columns = "Survived",axis=1)
X
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 25)
```

#### Modelling

```python
lr = LogisticRegression()
lr.fit(X_train,y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

#### Model2 Based on DecisionTreeClassifier

```python
dtc  = DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
y_pred = dtc.predict(X_test)
accuracy_score(y_test,y_pred)
```
